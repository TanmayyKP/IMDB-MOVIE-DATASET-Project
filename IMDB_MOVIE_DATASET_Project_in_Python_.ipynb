#Counting number of movies for each Genre

```python
import pandas as pd
from collections import Counter

# Read the CSV file into a DataFrame
data = pd.read_csv('/content/IMDB-Movie-Data.csv')

# Split and count occurrences
genre_counts = Counter()
for genres in data['Genre']:
    genre_counts.update(genres.split(','))  # Split and count occurrences

# Convert to dictionary or display as-is
print(dict(genre_counts))  # To see counts in dictionary format
```
